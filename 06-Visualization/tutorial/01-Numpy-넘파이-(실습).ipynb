## 배열 참조

- 배열을 슬라이싱 한 후 결과 값은 원본 배열을 참조(reference)합니다.
- 따라서, 슬라이싱한 배열 수정시 **원본 배열에서도 수정되는 현상이 발생**합니다.

```python
a = np.arange(1, 13)
a
```

슬라이싱하여 a1 변수에 대입합니다.

```python
# 슬라이싱으로 a1 생성
a1 = a[:5]
a1
```

`a1` 배열의 0번 인덱스의 값을 999로 변경합니다.

```python
a1[0] = 999
a1
```

```python
# 값을 참조하기 때문에 원본 데이터도 수정된다.
print(f'a  : {a}')
print(f'a1 : {a1}')
```

**방법1) 값을 참조하는 현상을 방지하려면 `copy() 사용`**

- `copy()`는 배열에 대한 복사본을 생성합니다.

```python
a = np.arange(1, 13)
a
```

```python
a1 = a[:5].copy()
```

```python
a1[0] = 999
a1
```

```python
# 결과 확인
print(f'a  : {a}')
print(f'a1 : {a1}')
```

**방법2) `fancy indexing` 사용**

```python
a = np.arange(1, 13)
a
```

```python
a1 = a[np.arange(5)]
a1
```

```python
a1[0] = 999
a1
```

```python
print(f'a  : {a}')
print(f'a1 : {a1}')
```

**(주의) 배열 연산시 배열 참조 현상이 발생 할 수 있습니다.**

[참고] 

- arr = arr * 10은 arr * 10 과정에서 새로운 배열을 생성합니다.

case 2) 곱셈 연산 수행

### var(), std(): 분산, 표준편차

- 분산과 표준편차: 데이터가 평균으로부터 얼마나 퍼져있는지 정도를 나타내는 지표

**분산($\sigma^2$) 공식**

$\Large \sigma^2 = \frac{\sum_{i=1}^{n}(x_i - \mu)^2} {n}$

**표준편차($\sigma$) 공식**

$\Large \sigma = \sqrt{\frac{\sum_{i=1}^{n}(x_i - \mu)^2} {n}}$

```python
# 샘플 배열 생성
arr = np.arange(1, 13).reshape(3, 4)
arr
```

```python
arr.var()
```

```python
arr.var(axis=0)
```

```python
arr.var(axis=1)
```

표준편차는 분산에 루트를 씌운 결과 입니다.

```python
np.sqrt(arr.var()), arr.std()
```

## 입출력

### save(): 단일 배열 저장

- 저장할 파일명, 배열(array) 순으로 저장합니다.
- 저장한 파일은 `.npy` 확장자로 저장됩니다. (파일로 저장)

### load(): 저장한 배열 로드

- `.npy`혹은 `.npz` 확장자를 가지는 저장된 배열을 불러옵니다.

```python
loaded_arr = np.load('my_arr.npy')
```

```python
loaded_arr
```

### savez(): 다중 배열 저장

- 다중 배열 저장시 `key=저장할배열` 형태로 저장합니다.
- 확장자는 `.npz`입니다.

```python
arr1 = np.arange(10)
arr2 = np.arange(10, 20)
```

```python
arr1
```

```python
arr2
```

키=배열 로 지정하여 저장합니다.

```python
np.savez('my_arrz', a1=arr1, a2=arr2)
```

저장한 배열을 불러옵니다.

```python
loaded_arr = np.load('my_arrz.npz')
```

`files`: 배열이 저장된 key 값을 조회할 수 있습니다.

```python
loaded_arr.files
```

key 값으로 배열을 불러옵니다.

```python
loaded_arr['a1']
```

```python
loaded_arr['a2']
```
